 <b>Задание 1а.</b> Создать Series из строчных букв латинского алфавита. В качестве индекса использовать порядковый номер буквы в алфавите, начиная с 1.



```python
import string
```

```python
s = np.array(list(string.ascii_lowercase)) # массив всех латинских заглавных букв
s_upper = np.array(list(string.ascii_uppercase)) # массив всех латинских заглавных букв

alpha = pd.Series(s, index=[i for i in range(1,len(s)+1)])
alpha_upper = pd.Series(s_upper, index=[i for i in range(1,len(s)+1)])
print(alpha)
```

 
 <b>Задание 1б.</b> Создать DataFrame, содержащий две колонки и состоящий из строчных и заглавных букв латинского алфавита. В качестве индекса использовать порядковый номер буквы в алфавите, начиная с 1. 

```python
data = pd.DataFrame(alpha, columns=['alpha'])
data['ALPHA'] = alpha_upper
data.head(5)

```

 <b>Задание 2а.</b> С клавиатуры вводится строка, состоящая из латинских букв и пробелов. Создать словарь, ключами которого являются буквы, входящие во введённую строку (причем буквы 'a' и 'A' являются разными ключами), а значениями $-$ количество вхождений буквы в строку. 
 


```python
user_string = input('>>> ')
set_alpha = set()
count_alphas = {}
for alpha in user_string:

    set_alpha.add(alpha.lower())#для 2б

    if alpha == " ":
        continue

    if alpha in count_alphas:
        count_alphas[alpha]+=1
    else:
        count_alphas[alpha]=1
        
```

```python
count_alphas
```

 <b>Задание 2б.</b> Оставить в DataFrame, созданном в <b>Задании 1б.</b> только строки только букв, являющиеся ключами словаря (оставить строку нужно, если в тексте появилась хотя бы одна строчная и/или заглавная буква). 

```python
new_data = data[data['alpha'].isin(set_alpha)]
new_data
```

 <b>Задание 2в.</b> Добавить в DataFrame новую колонку, значениями которой является число вхождений в строку соответствующей буквы.

```python
counter = []
for item in new_data['alpha']:
    if item in count_alphas.keys():
      counter.append(count_alphas[item])
    elif item.upper() in count_alphas.keys():
      counter.append(count_alphas[item.upper()])
      
new_data['count'] = counter
```

<b>Задание 3.</b> Дан DataFrame DF3, содержащий случайные числа от 1 до 100. Добавить в него четыре колонки: максимальное значение в строке, минимальное значение в строке, среднее арифметическое элементов строки, сумму элементов строки. Имена колонок _max_, _min_, _mean_ и _sum_ соответственно.


```python
DF3 = pd.DataFrame(np.random.randint(-100, 100, 100).reshape(10, -1), columns=list(range(10)), index=list('abcdefghij'))
DF3
```

```python
sum3 = DF3.sum(axis=1)
max3 = DF3.max(axis=1)
min3 = DF3.min(axis=1)
mean3 = DF3.mean(axis=1)

DF3['max'] = max3
DF3['min'] = min3
DF3['sum'] = sum3
DF3['mean'] = mean3
DF3
```

<b>Задание 4*.</b> Добавить колонку _nearest_ в DataFrame DF3. В качестве значения для каждой строки указать строку с наиболее близким средним значением элементов в строке. 

```python
DF3.loc['a']['mean']
```

```python
nears_vector = []
for index, row in DF3.iterrows():
    diff = abs(DF3['mean'] - DF3.loc[index]['mean'])
    diff = diff.sort_values(axis=0)
    #print(diff)
    more_nearest = diff[1:].idxmin()

    nears_vector.append(more_nearest)

DF3['nears'] = nears_vector
DF3
```
